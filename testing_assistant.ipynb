```python
from IPython.display import Audio, display, Javascript
import numpy as np
from scipy.io import wavfile
import speech_recognition as sr
import wikipedia
import wolframalpha
import datetime
import os
import sys
from gtts import gTTS
import librosa
import time
import requests
import re
from google.colab import files

# Text-to-Speech function with sass
def speak(to_speak="Well, what do you want now?"):
    tts = gTTS(to_speak)
    audio_path = 'out.wav'  # Colab’s temp directory
    tts.save(audio_path)
    display(Audio(audio_path, autoplay=True))
    time.sleep(librosa.get_duration(filename=audio_path))
    if os.path.exists(audio_path):
        os.remove(audio_path)
    return 200

# Speech recognition with better feedback
def _recognize(audio_file):
    r = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio_data = r.record(source)
        try:
            text = r.recognize_google(audio_data)
            return text.lower()
        except sr.UnknownValueError:
            speak("Sorry, I couldn’t understand that. Speak like you mean it next time.")
            return "None"
        except sr.RequestError as e:
            speak("My speech recognition is on the fritz. Blame the internet.")
            return "None"

# Audio upload function for Colab
def takeCommand():
    print("Upload a .wav file with your command.")
    uploaded = files.upload()
    if uploaded:
        audio_file = list(uploaded.keys())[0]
        text_recognized = _recognize(audio_file)
        print(f"You said: {text_recognized}")
        if os.path.exists(audio_file):
            os.remove(audio_file)  # Clean up
        return text_recognized
    else:
        speak("No file? Fine, type it then, lazy.")
        query = input("Type your command: ")
        return query.lower()

# Open web pages with attitude
def open_web(url, message="Here’s your precious link."):
    speak(message)
    display(Javascript(f'window.open("{url}");'))

# Simple web search fallback
def web_search(query):
    url = f"https://www.google.com/search?q={query}"
    open_web(url, f"Searching for '{query}'. Hope it’s worth my time.")

# Greeting with personality
def wishMe():
    hour = int(datetime.datetime.now().hour)
    if 0 <= hour < 12:
        speak("Morning, sunshine! Ready to boss me around?")
    elif 12 <= hour < 18:
        speak("Afternoon, huh? Don’t expect me to make you lunch.")
    else:
        speak("Evening already? Time flies when I’m doing all the work.")
    name = "Parhsu the Magnificent"
    speak(f"I’m {name}, your oh-so-smart assistant.")

# Set username with sass
def usrname():
    speak("What’s your name, mortal?")
    uname = takeCommand()
    if uname == "none":
        uname = "Anonymous"
        speak("Fine, I’ll just call you Anonymous. How original.")
    else:
        speak(f"Welcome, {uname}. Try not to bore me.")

# Smart assistant logic
def smart_bot():
    wishMe()
    usrname()
    while True:
        speak("What’s on your tiny mind now?")
        query = takeCommand()
        print(f"Query: {query}")

        # Wikipedia search
        if 'wikipedia' in query:
            speak("Digging through Wikipedia. Hold your horses.")
            query = query.replace("wikipedia", "").strip()
            try:
                results = wikipedia.summary(query, sentences=3)
                speak("Here’s the short version, because you’re clearly too busy to read.")
                print(results)
                speak(results)
            except wikipedia.exceptions.DisambiguationError:
                speak("Too many options. Be more specific, genius.")
            except wikipedia.exceptions.PageError:
                speak("Nothing found. Maybe it doesn’t exist, or you’re just bad at this.")

        # Web openings
        elif 'open youtube' in query:
            open_web("https://www.youtube.com", "YouTube, because staring at cat videos is peak productivity.")

        elif 'open google' in query:
            open_web("https://www.google.co.in", "Google. Go figure it out yourself for once.")

        elif 'open stack overflow' in query:
            open_web("https://www.stackoverflow.com", "Stack Overflow. Good luck pretending you’re a coder.")

        # Time/Date queries
        elif 'time' in query:
            current_time = datetime.datetime.now().strftime("%H:%M:%S")
            speak(f"It’s {current_time}. You got a watch, or just too lazy to check?")

        elif 'date' in query or 'day' in query:
            current_date = datetime.datetime.now().strftime("%B %d, %Y")
            speak(f"Today’s {current_date}. Mark it on your calendar if you can find it.")

        # Simple calculations


        # How are you
        elif 'how are you' in query:
            speak("I’m fantastic, thanks for asking. You? Oh wait, I don’t really care.")

        # Search anything else
        elif 'search' in query or 'what is' in query:
            query = query.replace("search", "").replace("what is", "").strip()
            web_search(query)

        # Exit
        elif 'exit' in query or 'bye' in query:
            speak("Finally, some peace. Catch you later, or not.")
            break
        elif 'calculate' in query or 'what is' in re.search(r'\d+[\+\-\*/]\d+', query):
            expr = re.search(r'\d+[\+\-\*/]\d+', query)
            if expr:
                try:
                    result = eval(expr.group())  # Basic arithmetic only
                    speak(f"Your little math problem equals {result}. Impressed yet?")
                except:
                    speak("That’s not math, that’s gibberish. Try again.")
            else:
                speak("Give me something to calculate, not just hot air.")
        # Default sass
        else:
            speak("I heard you, but I’ve got no clue what you’re on about. Try harder.")
            web_search(query)

# Run the smart bot
smart_bot()
```
